```python
# Add your imports here
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
%matplotlib inline
```

```python
persons = pd.read_csv("persons_occurences_day.csv").dropna()
persons = persons.sort_values(['Year', 'Month', 'Day', 'Count'], ascending=[True, True, True, False])
persons.Year = persons.Year.astype(int)
persons.Month = persons.Month.astype(int)
persons.Day = persons.Day.astype(int)
persons['Date'] = persons.apply(lambda row: datetime(row['Year'], row['Month'], row['Day']), axis=1)
persons = persons.drop(columns = ['Day', 'Month', 'Year'])
persons = persons.set_index('Date')
```
